<summary style="font-size: 21px; font-weight: 600;">Deductive reasoning</summary>

Deductive reasoning is a form of logical reasoning in which a conclusion necessarily follows from given premises.

Core definition

If the premises are true and the reasoning is valid, the conclusion must be true—there is no uncertainty.

Basic structure

Deductive reasoning moves from:
general principles → specific conclusions

Example:

All mammals are warm-blooded.

Whales are mammals.

Therefore, whales are warm-blooded.

Key characteristics

Necessity: conclusions are guaranteed by logic

Validity vs. truth:

Valid: the conclusion follows logically

Sound: valid and premises are true

Formal structure: often expressed as syllogisms

Deductive vs. other reasoning

Deductive: certainty (if premises are true)

Inductive: probability (generalizing from examples)

Abductive: best explanation (inference to the most likely cause)

Common uses

Mathematics and formal logic

Law and legal arguments

Philosophy and ethics

Computer science (program logic, proofs)

In one sentence

Deductive reasoning derives conclusions that must be true if the premises are true.

</details>
